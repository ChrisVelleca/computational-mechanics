```python
import tf_idf.core as tf_idf
import nltk
nltk.download('punkt_tab')
```

```python
AI = '''To properly brush your teeth, start by selecting a soft-bristled toothbrush and applying a pea-sized amount of fluoride toothpaste. Hold the toothbrush at a 45-degree angle to your gumline to ensure effective plaque removal. Using gentle pressure, move the brush in short, circular motions or small back-and-forth strokes at a rate of approximately 100 strokes per minute. Spend at least two minutes brushing in total, devoting about three to five seconds per tooth to ensure thorough cleaning. Be sure to cover all 32 teeth (or 28 if wisdom teeth are removed), including the front, back, and chewing surfaces. Proper technique involves slightly deflecting the bristles against the tooth surface and gumline to allow for optimal cleaning without causing damage to the gums. A toothbrush should generally be replaced every three to four months or sooner if the bristles become frayed, as worn bristles are less effective in removing plaque and debris.'''
compare = tf_idf.preprocess_text(AI)
```

```python
ME = '''To properly brush your teeth, start by selecting a soft-bristled toothbrush and applying a pea-sized amount of toothpaste. Hold the toothbrush at a 45-degree angle to your gumline to ensure effective plaque removal. Using gentle pressure, move the brush in short, circular motions or small back-and-forth strokes at a rate of approximately 100 strokes per minute. You should spend at least two minutes brushing in total, devoting about three to five seconds per tooth to ensure thorough cleaning. Be sure to cover all 32 teeth, including the front, back, and chewing surfaces. Proper technique involves slightly deflecting the bristles against the tooth surface and gumline to allow for optimal cleaning without causing damage to the gums. A toothbrush should generally be replaced every three to four months or sooner if the bristles become frayed, as worn bristles are less effective in removing plaque and debris.'''
compare = pd.concat([compare, tf_idf.preprocess_text(ME)],
                    ignore_index=True)
compare
```

```python
tf_idf.cosineSimilarity(compare)
```
